```python
from tensorflow import keras
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
import sklearn.metrics as plot_metric
import sklearn.metrics as accuracy_score
```

```python
# load dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print('x_train: {}'.format(x_train.shape))
print('y_train: {}'.format(y_train.shape))
print('x_test: {}'.format(x_test.shape))
print('y_test: {}'.format(y_test.shape))
```

```python
# Convert class vectors to binary class matrices.
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
print('y_train: {}'.format(y_train.shape))
print('y_test: {}'.format(y_test.shape))
```

```python
#normalization

x_train = x_train.astype(float) / 255.
x_test  = x_test.astype(float) / 255.
```

```python
# define model
def define_model(shape, num_layers):
    model = keras.Sequential()
    model.add(keras.layers.Input(shape=shape))
    model.add(keras.layers.Flatten())
    for l in range(num_layers):
        model.add(keras.layers.Dense(units=50, activation='relu'))
    model.add(keras.layers.Dense(units=num_classes, activation='sigmoid'))

    return model
```

```python
# compile model
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

# train model
history = model.fit(x_train, y_train,
                    batch_size=512,
                    epochs=1000,
                    validation_data=(x_test, y_test),
                    shuffle=True)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper right')
```
